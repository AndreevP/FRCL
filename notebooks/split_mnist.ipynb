```python
import sys
sys.path.append("..")
# imports pipelines for different models
from frcl.experiments.pipeline import BaselineTrainDemo, FRCLTrainDemo, FRCLStatTrace
from frcl.tasks import SplitMnistTask, PermutedMnistTask
import seaborn as sns
import numpy as np
sns.set()
```

```python
# create split_mnist Task suite
task = SplitMnistTask()
```

### Baseline experiments with different settings

#### Experiment with replay buffer of size 2

```python
# creates and runs baseline model with n_inducing = 2
baseline_pipeline = BaselineTrainDemo(task, n_repeat=1, device='cpu', n_epochs=4, lr=1e-5, batch_size=100, n_inducing=2)
res = baseline_pipeline.run()
```

```python
# shows results for the launch
mean_accuracy = np.mean(res[0][0].get_solver_estimations(len(task))[1])
res[0][0].draw_task_estimations(len(task))
print('Mean accuracy: ', round(mean_accuracy, 3))
```

#### Experiment with replay buffer of size 40

```python
baseline_pipeline = BaselineTrainDemo(task, n_repeat=1, device='cpu', n_epochs=4, lr=1e-5, batch_size=100, n_inducing=40)
res = baseline_pipeline.run()
```
